```python
import csv
f = open("seoul.csv")
data = csv.reader(f)
next(data)

dif = 0
mx = 0
mn = 0
dif_date = ''
mx_date = ''
mn_date = ''

for row in data:
    high = float(row[-1])
    low = float(row[-2])
    if(high - low > dif):
        dif = high - low
        dif_date = row[0]
    if(low < mn):
        mn = low
        mn_date = row[0]
    if(high > mx):
        mx = high
        mx_date = row[0]
        
print("일교차가 가장 큰 날")
print("날짜 : {0:s}" .format(dif_date))
print("최저기온이 가장 낮은 날")
print("날짜 : {0:s}" .format(mn_date))
print("최고기온이 가장 높은 날")
print("날짜 : {0:s}" .format(mx_date))
```

```python
import csv
# 20년 12월 노선별 역별 이용현황
f = open("subwayfee.csv")
data = csv.reader(f)
next(data)

line = ''
mx = {}
mn = {}

# 사용월, 호선명, 역ID, 지하철역, 승차승객수, 하차승객,,
for row in data:
    if(line != row[1]):
        if(len(mx) != 0):
            stn_list = list(mx.keys())
            cnt_list = list(mx.values())
            sorted_cnt_li = list(mx.values())
            sorted_cnt_li.sort(reverse=True)
            print("<", line, ">")
            print("이용객이 가장 많은 역 순위")
            for i in range(len(sorted_cnt_li)):
                for j in range(len(cnt_list)):
                    if(sorted_cnt_li[i] == cnt_list[j]):
                        print("{0:d}. {1:s} ({2:d}명)" .format((i + 1), stn_list[j], cnt_list[j]))
            stn_list = list(mn.keys())
            cnt_list = list(mn.values())
            sorted_cnt_li = list(mn.values())
            sorted_cnt_li.sort()
            print("이용객이 가장 적은 역 순위")
            for i in range(len(sorted_cnt_li)):
                for j in range(len(cnt_list)):
                    if(sorted_cnt_li[i] == cnt_list[j]):
                        print("{0:d}. {1:s} ({2:d}명)" .format((i + 1), stn_list[j], cnt_list[j]))
            print()
        line = row[1]
        mx.clear()
        mn.clear()
    row[-2] = int(row[-2].replace(",", ""))   #하차승객수
    row[-3] = int(row[-3].replace(",", ""))   #승차승객수
    count = row[-2] + row[-3]
    mx[row[3]] = count   #row[3]은 역명
    mn[row[3]] = count
    if(len(mx) > 3):
        count_list = list(mx.values())
        sorted_cnt_li = list(mx.values())
        sorted_cnt_li.sort()
        min_cnt = sorted_cnt_li[0]
        for i in range(3, -1, -1):
            if(min_cnt == count_list[i]):
                rm_key = list(mx.keys())[i]
                del mx[rm_key]
    if(len(mn) > 3):
        count_list = list(mn.values())
        sorted_cnt_li = list(mn.values())
        sorted_cnt_li.sort()
        max_cnt = sorted_cnt_li[-1]
        for i in range(3, -1, -1):
            if(max_cnt == count_list[i]):
                rm_key = list(mn.keys())[i]
                del mn[rm_key]
        
stn_list = list(mx.keys())
cnt_list = list(mx.values())
sorted_cnt_li = list(mx.values())
sorted_cnt_li.sort(reverse=True)

print("<", line, ">")
print("이용객이 가장 많은 역 순위")
for i in range(len(sorted_cnt_li)):
    for j in range(len(cnt_list)):
        if(sorted_cnt_li[i] == cnt_list[j]):
            print("{0:d}. {1:s} ({2:d}명)" .format((i + 1), stn_list[j], cnt_list[j]))
stn_list = list(mn.keys())
cnt_list = list(mn.values())
sorted_cnt_li = list(mn.values())
sorted_cnt_li.sort()
print("이용객이 가장 적은 역 순위")
for i in range(len(sorted_cnt_li)):
    for j in range(len(cnt_list)):
        if(sorted_cnt_li[i] == cnt_list[j]):
            print("{0:d}. {1:s} ({2:d}명)" .format((i + 1), stn_list[j], cnt_list[j]))
```
