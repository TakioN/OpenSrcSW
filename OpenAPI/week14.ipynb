```python
import csv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc
import matplotlib.patches as mpatches

f = open("population.csv")
data = csv.reader(f)
next(data)
next(data)
target = []  #광진구 제외 인구 수 담을 리스트
var = []     #광진구 제외 분산 담을 리스트
isFirstLine = True    #첫 줄 여부 판단 
max_idx = 0  #광진구와 차이가 가장 큰 도시의 인덱스
min_idx = 0  #광진구와 차이가 가장 작은 도시의 인덱스

for line in data:
    if("광진구" in line[0]):
        target = [line[0]] + line[3:104]
        target_arr = np.array(target)
        continue
  
    line = [line[0]] + line[3:104]  #구 이름 + 인구수
    if(isFirstLine == False):       
        a = np.array([line])
        city = np.append(city, a, axis=0)
    else:
        city = np.array([line])
        isFirstLine = False
for ct in city:
    ct_var = 0
    for i in range(1, len(target_arr)):
        ct[i] = ct[i].replace(",", "")
        target_arr[i] = target_arr[i].replace(",", "")
        ct_var += (int(ct[i]) - int(target_arr[i]))**2  #구 별 분산
    var.append(ct_var)

#폰트 지정
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc("font", family=font_name)
#

max_idx = var.index(max(var))
min_idx = var.index(min(var))
max_value = city[max_idx][1:].astype(np.int32)  #구 이름을 제외한 인구수만 정수로 저장
min_value = city[min_idx][1:].astype(np.int32)  #구 이름을 제외한 인구수만 정수로 저장
tar_value = target_arr[1:].astype(np.int32)     #구 이름을 제외한 인구수만 정수로 저장
std_max = (var[max_idx] / 101) ** (1/2)         #차이가 가장 큰 구의 표준편차
std_min = (var[min_idx] / 101) ** (1/2)         #차이가 가장 작은 구의 표준편차

plt.style.use("ggplot")
plt.plot(max_value, label=city[max_idx][0] + "(" + str(std_max) + ")")
plt.plot(min_value, label=city[min_idx][0] + "(" + str(std_min) + ")")
plt.plot(tar_value, label=target_arr[0])
plt.legend(loc="upper right")
plt.show()
          
f.close()
```
